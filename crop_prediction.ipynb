## hybrid algorithm

```python
from sklearn.ensemble import StackingRegressor, RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
import numpy as np

# Reduced complexity base models
base_models = [
    ('random_forest', RandomForestRegressor(n_estimators=11, max_depth=5, n_jobs=-1, random_state=42)),
    ('decision_tree', DecisionTreeRegressor(max_depth=5, random_state=42))
]

meta_learner = LinearRegression()

stacking_regressor = StackingRegressor(estimators=base_models, final_estimator=meta_learner, n_jobs=-1)

y_train_flattened = np.ravel(y_train)

stacking_regressor.fit(x_train, y_train_flattened)

y_pred = stacking_regressor.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error (MAE):", mae)
```

```python
from sklearn.metrics import r2_score

# Calculate R-squared score
r2 = r2_score(y_test, y_pred)
print("R-squared score:", r2)
```
